### Popcorn Hack 1

```python
# created my own sample data as you guys didn't provide any
student_data = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David'],
    'score': [85, 92, 78, 88]
})

def find_students_in_range(df, min_score, max_score):
    return df[(df['score'] >= min_score) & (df['score'] <= max_score)]

# testing the function
print(find_students_in_range(student_data, 80, 90))
```

### Popcorn Hack 2

```python
def add_letter_grades(df):
    def get_letter(score):
        if score >= 90:
            return 'A'
        elif score >= 80:
            return 'B'
        elif score >= 70:
            return 'C'
        elif score >= 60:
            return 'D'
        else:
            return 'F'
    
    df['Letter'] = df['score'].apply(get_letter)
    return df

add_letter_grades(student_data)
```

### Popcorn Hack 3

```python
def find_mode(series):
    return series.mode().iloc[0]

print(find_mode(pd.Series([1, 2, 2, 3, 4, 2, 5])))
```

**Questions I Attempted to Answer:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../../assets/fire.csv')

print("Dataset Overview:")
print(data.head())

print("\nDataset Information:")
print(data.info())

print("\nSummary Statistics:")
print(data.describe())

print("\nMissing Values:")
print(data.isnull().sum())

print("\nColumns:", data.columns.tolist())


print("\nUnique satellites:", data.satellite.unique())
print("Unique daynight indicators:", data.daynight.unique())
print("Date range:", data.acq_date.min(), "to", data.acq_date.max())


analysis_data = data.copy()


analysis_data['acq_date'] = pd.to_datetime(analysis_data['acq_date'])



plt.figure(figsize=(10, 6))
sns.scatterplot(x='confidence', y='frp', hue='daynight', data=analysis_data)
plt.title('Fire Radiative Power vs. Detection Confidence')
plt.xlabel('Confidence Level')
plt.ylabel('Fire Radiative Power (FRP)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('frp_vs_confidence.png')
plt.close()


plt.figure(figsize=(10, 6))
sns.boxplot(x='daynight', y='frp', data=analysis_data)
plt.title('Fire Radiative Power Distribution: Day vs Night')
plt.xlabel('Time of Day')
plt.ylabel('Fire Radiative Power (FRP)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('frp_by_daynight.png')
plt.close()


plt.figure(figsize=(10, 6))
sns.scatterplot(x='brightness', y='frp', hue='daynight', data=analysis_data)
plt.title('Relationship between Brightness and Fire Radiative Power')
plt.xlabel('Brightness Temperature (K)')
plt.ylabel('Fire Radiative Power (FRP)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('brightness_vs_frp.png')
plt.close()


plt.figure(figsize=(12, 8))
sns.scatterplot(x='longitude', y='latitude', size='frp', hue='confidence', 
                sizes=(20, 200), data=analysis_data)
plt.title('Fire Hotspot Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('fire_hotspot_map.png')
plt.close()


analysis_data['lat_bin'] = pd.cut(analysis_data['latitude'], bins=5)
analysis_data['lon_bin'] = pd.cut(analysis_data['longitude'], bins=5)

region_frp = analysis_data.groupby(['lat_bin', 'lon_bin'])['frp'].mean().reset_index()
print("\nAverage FRP by geographical region:")
print(region_frp.sort_values('frp', ascending=False).head(10))


satellite_frp = analysis_data.groupby('satellite')['frp'].agg(['mean', 'median', 'count']).reset_index()
print("\nAverage FRP by satellite:")
print(satellite_frp.sort_values('mean', ascending=False))

plt.figure(figsize=(10, 6))
sns.barplot(x='satellite', y='mean', data=satellite_frp)
plt.title('Average Fire Radiative Power by Satellite')
plt.xlabel('Satellite')
plt.ylabel('Average FRP')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('avg_frp_by_satellite.png')
plt.close()


high_temp_threshold = 320
high_temp_count = analysis_data[analysis_data['brightness'] > high_temp_threshold].shape[0]
total_count = analysis_data.shape[0]
high_temp_percentage = (high_temp_count / total_count) * 100

print(f"\nPercentage of fire incidents with brightness temperature above {high_temp_threshold}K: {high_temp_percentage:.2f}%")


plt.figure(figsize=(12, 6))
sns.histplot(data=analysis_data, x='frp', hue='daynight', bins=30, kde=True)
plt.title('Distribution of Fire Radiative Power')
plt.xlabel('Fire Radiative Power (FRP)')
plt.ylabel('Count')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('frp_distribution.png')
plt.close()


analysis_data['hour'] = analysis_data['acq_time'].apply(lambda x: int(str(x).zfill(4)[:2]))
hourly_fires = analysis_data.groupby('hour').size().reset_index(name='count')

plt.figure(figsize=(12, 6))
sns.barplot(x='hour', y='count', data=hourly_fires)
plt.title('Fire Detections by Hour of Day')
plt.xlabel('Hour (24-hour format)')
plt.ylabel('Number of Detections')
plt.xticks(range(0, len(hourly_fires)), hourly_fires['hour'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('fires_by_hour.png')
plt.close()


correlation = analysis_data['brightness'].corr(analysis_data['confidence'])
print(f"\nCorrelation between brightness and confidence: {correlation:.4f}")


print("\nSUMMARY OF FINDINGS:")
print("1. Geographical Distribution: The data shows fire hotspots in several regions, with varying intensities.")
print(f"2. Detection Conditions: {satellite_frp.iloc[0]['satellite']} satellite recorded the highest average FRP at {satellite_frp.iloc[0]['mean']:.2f}.")
print(f"3. High Temperature Incidents: {high_temp_percentage:.2f}% of fire incidents recorded brightness temperatures above {high_temp_threshold}K.")
print(f"4. Brightness-Confidence Correlation: There is a {correlation:.4f} correlation between fire brightness and detection confidence.")
print(f"5. Day/Night Pattern: The data shows {'higher' if analysis_data[analysis_data['daynight']=='D']['frp'].mean() > analysis_data[analysis_data['daynight']=='N']['frp'].mean() else 'lower'} average FRP during daylight hours compared to nighttime.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


plt.style.use('seaborn-v0_8')
sns.set(font_scale=1.2)


data = pd.read_csv('../../assets/fire.csv')


data['acq_date'] = pd.to_datetime(data['acq_date'])
data['hour'] = data['acq_time'].apply(lambda x: int(str(x).zfill(4)[:2]))

import os
if not os.path.exists('visualizations'):
    os.makedirs('visualizations')


plt.figure(figsize=(12, 8))
sns.scatterplot(x='confidence', y='frp', hue='daynight', data=data, alpha=0.7)
plt.title('Fire Radiative Power vs Detection Confidence', fontsize=16)
plt.xlabel('Confidence Level', fontsize=14)
plt.ylabel('Fire Radiative Power (FRP)', fontsize=14)
plt.legend(title='Day/Night', fontsize=12, title_fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/1_frp_vs_confidence.png', dpi=300)
plt.close()
print("1. Created visualization: Fire Radiative Power vs Detection Confidence")


plt.figure(figsize=(10, 8))
sns.boxplot(x='daynight', y='frp', data=data)
plt.title('Fire Radiative Power Distribution: Day vs Night', fontsize=16)
plt.xlabel('Time of Day', fontsize=14)
plt.ylabel('Fire Radiative Power (FRP)', fontsize=14)
plt.xticks([0, 1], ['Night (N)', 'Day (D)'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/2_day_night_frp_comparison.png', dpi=300)
plt.close()
print("2. Created visualization: Fire Radiative Power Distribution: Day vs Night")


plt.figure(figsize=(12, 8))
g = sns.scatterplot(x='brightness', y='frp', hue='daynight', size='confidence', 
              sizes=(20, 200), data=data, alpha=0.7)
plt.title('Brightness Temperature vs Fire Radiative Power', fontsize=16)
plt.xlabel('Brightness Temperature (K)', fontsize=14)
plt.ylabel('Fire Radiative Power (FRP)', fontsize=14)
plt.legend(title='Day/Night', fontsize=12, title_fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/3_brightness_vs_frp.png', dpi=300)
plt.close()
print("3. Created visualization: Brightness Temperature vs Fire Radiative Power")


plt.figure(figsize=(14, 10))
sns.scatterplot(x='longitude', y='latitude', size='frp', hue='confidence', 
                sizes=(20, 200), palette='viridis', data=data, alpha=0.7)
plt.title('Geographical Distribution of Fire Hotspots', fontsize=16)
plt.xlabel('Longitude', fontsize=14)
plt.ylabel('Latitude', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/4_fire_hotspot_map.png', dpi=300)
plt.close()
print("4. Created visualization: Geographical Distribution of Fire Hotspots")

data['lat_bin'] = pd.cut(data['latitude'], bins=5)
data['lon_bin'] = pd.cut(data['longitude'], bins=5)


region_frp = data.groupby(['lat_bin', 'lon_bin'])['frp'].agg(['mean', 'count']).reset_index()
region_frp = region_frp.sort_values('mean', ascending=False)


print("\n5. Top regions by average fire intensity (FRP):")
print(region_frp.head(5))


satellite_frp = data.groupby('satellite')['frp'].agg(['mean', 'median', 'count']).reset_index()
print("\n6. Average fire intensity by satellite:")
print(satellite_frp)

plt.figure(figsize=(10, 6))
sns.barplot(x='satellite', y='mean', data=satellite_frp)
plt.title('Average Fire Radiative Power by Satellite', fontsize=16)
plt.xlabel('Satellite', fontsize=14)
plt.ylabel('Average FRP', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/6_avg_frp_by_satellite.png', dpi=300)
plt.close()
print("6. Created visualization: Average Fire Radiative Power by Satellite")


high_temp_threshold = 320
high_temp_count = data[data['brightness'] > high_temp_threshold].shape[0]
total_count = data.shape[0]
high_temp_percentage = (high_temp_count / total_count) * 100

print(f"\n7. Percentage of fire incidents with brightness temperature above {high_temp_threshold}K: {high_temp_percentage:.2f}%")


plt.figure(figsize=(10, 8))
plt.pie([high_temp_count, total_count - high_temp_count], 
        labels=[f'>{high_temp_threshold}K ({high_temp_percentage:.1f}%)', 
                f'â‰¤{high_temp_threshold}K ({100-high_temp_percentage:.1f}%)'],
        autopct='%1.1f%%', startangle=90, explode=(0.1, 0), 
        colors=['#ff7f0e', '#1f77b4'], shadow=True)
plt.title(f'Proportion of Fires with Brightness Temperature > {high_temp_threshold}K', fontsize=16)
plt.axis('equal')
plt.tight_layout()
plt.savefig('visualizations/7_high_temp_fires_percentage.png', dpi=300)
plt.close()
print("7. Created visualization: Proportion of High Temperature Fires")


hourly_fires = data.groupby('hour').size().reset_index(name='count')

plt.figure(figsize=(14, 8))
sns.barplot(x='hour', y='count', data=hourly_fires)
plt.title('Fire Detections by Hour of Day', fontsize=16)
plt.xlabel('Hour (24-hour format)', fontsize=14)
plt.ylabel('Number of Detections', fontsize=14)
plt.xticks(range(len(hourly_fires)), hourly_fires['hour'])
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/8_fires_by_hour.png', dpi=300)
plt.close()
print("8. Created visualization: Fire Detections by Hour of Day")


plt.figure(figsize=(14, 8))
sns.histplot(data=data, x='frp', hue='daynight', bins=30, kde=True)
plt.title('Distribution of Fire Radiative Power', fontsize=16)
plt.xlabel('Fire Radiative Power (FRP)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('visualizations/9_frp_distribution.png', dpi=300)
plt.close()
print("9. Created visualization: Distribution of Fire Radiative Power")


correlation = data['brightness'].corr(data['confidence'])
print(f"\n10. Correlation between brightness and confidence: {correlation:.4f}")


correlation_vars = ['brightness', 'confidence', 'frp', 'bright_t31']
correlation_matrix = data[correlation_vars].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Key Fire Variables', fontsize=16)
plt.tight_layout()
plt.savefig('visualizations/10_correlation_heatmap.png', dpi=300)
plt.close()
print("10. Created visualization: Correlation Matrix of Key Fire Variables")

print("\nAll visualizations have been saved to the 'visualizations' directory.")
print("\nSummary of questions addressed in this analysis:")
print("1. Is there a relationship between fire detection confidence and fire intensity (FRP)?")
print("2. How does fire intensity (FRP) differ between day and night observations?")
print("3. What is the relationship between brightness temperature and fire radiative power?")
print("4. Where are the fire hotspots geographically distributed?")
print("5. Do different geographical regions show different patterns of fire intensity?")
print("6. Which satellite observed higher average fire intensities?")
print("7. What percentage of fire incidents recorded high brightness temperatures?")
print("8. How are fire detections distributed throughout the day?")
print("9. What is the distribution pattern of fire radiative power?")
print("10. Is there a correlation between brightness temperature and detection confidence?")
```
